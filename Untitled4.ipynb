
# atividade 1 

```python
# Entradas de dados
idade = int(input("Digite sua idade: "))

# Primeiro filtro: Idade
if idade >= 18:
    tem_carteira = input("Você tem carteira de motorista? (sim/nao): ")
    
    # Segundo filtro: Só entra aqui se o primeiro for verdadeiro
    if tem_carteira == "sim":
        print("Tudo certo! Você pode dirigir.")
    else:
        print("Você tem idade, mas não tem o documento. Não pode dirigir.")
        
else:
    # Se não passou nem no primeiro filtro
    print("Você é menor de idade. Não pode dirigir de jeito nenhum.")
```

```python
senha_mestra = "python"

while True:
    entrada = input("Digite a senha secreta: ")
    
    if entrada == senha_mestra:
        print("Parabéns! Você acertou.")
        break
    else:
        print("Errou! O loop continua...")
```

# exercico 6

```python
opcao = -1

while opcao != 0:
    print("\n--- MENU DE OPÇÕES ---")
    print("1. Perfil")
    print("2. Configurações")
    print("0. Sair")
    
    opcao = int(input("Escolha uma opção: "))
    
    if opcao == 1:
        print("Acessando Perfil...")
    elif opcao == 2:
        print("Acessando Configurações...")
    elif opcao == 0:
        print("Saindo do sistema...")
    else:
        print("Opção inválida!")
```
