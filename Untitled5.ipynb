```python
contar_pares = 0

for i in range(5):
    numero = int(input(f"Digite o {i+1}º número: "))
    if numero % 2 == 0:
        contar_pares += 1

print(f"Você digitou {contar_pares} números pares.")
```

```python
nota = -1  

while nota < 0 or nota > 10:
    nota = float(input("Digite uma nota entre 0 e 10: "))
    
    if nota < 0 or nota > 10:
        print("Valor inválido! Tente novamente.")

print(f"Nota {nota} registrada com sucesso!")
```

```python
numero = int(input("Digite um número: "))

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

```python
idade = int(input("Digite sua idade: "))
tem_documento = input("Possui documento oficial? (sim/nao): ").lower()

if idade >= 18 and tem_documento == "sim":
    print("Acesso autorizado.")
else:
    print("Acesso negado. É necessário ser maior de idade e portar documento.")
```

```python
usuario_correto = "admin"
senha_correta = "1234"
tentativas = 3

while tentativas > 0:
    user = input("Usuário: ")
    senha = input("Senha: ")

    if user == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso! Bem-vindo.")
        break
    else:
        tentativas -= 1
        if tentativas > 0:
            print(f"Dados incorretos. Você ainda tem {tentativas} tentativa(s).")
        else:
            print("Acesso bloqueado. Número de tentativas excedido.")
```
