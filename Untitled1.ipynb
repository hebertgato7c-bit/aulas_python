# Atividade 2 if e else

# exercico 1

```python
idade = int(input("Digite sua idade"))
```

```python
if idade >=12:
    ingresso = input("Você tem ingresso?) (s/n): ")
    
    if ingresso == "s":
        print("Entrada liberada")
        
    elif ingresso == "n":
        print("Compre o ingresso")
        
    else:
        print("acesso invalido")
else: 
    print("Idade insuficiente")
```

# exercico 7

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        if numero % 3 == 0:
            print(f"O número {numero} é múltiplo de 2 e de 3.")
        else:
            print(f"O número {numero} é múltiplo de 2, mas não de 3.")
    else:
        print(f"O número {numero} não é múltiplo de 2.")
else:
    print("Por favor, digite um número positivo.")
```

# exercicio 8

```python
nome = input("Digite seu nome: ").strip()

if nome != "":
    idade = int(input("Digite sua idade: "))
    
    if idade >= 18:
        print("Cadastro completo")
    else:
        print("Cadastro incompleto: menor de idade")
else:
    print("Nome obrigatório")
```
